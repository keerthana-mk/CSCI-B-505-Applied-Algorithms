Suppose you are given two sequences D 1 and D 2 of n elements, possibly containing duplicates,on which a total
order relation is defined. Describe an efficient algorithm for determining if D1 and D2 contain the same set 
of elements. What is the running time of this method?

## Solution

### Case1:

**If there are only unique elements in both the sequences:**
We can hash both the sequences and this happens in linear time O(n).

* First iterate through the sequence D1 and store it as a key in hashtable.
* Similarly iterate through the sequence D2 and store it as a key in hashtable.
* Check for every element in D2, if there is a entry in D1 hashtable.If any element does not find the entry D1 hashtable then teh sequences are not equal.

**Pseudocode:**
~~~
checkEqualSequenceHash(D1, D2)
 // initialise hashtables D1 and D2
    let D1= hashtable()
    let D2= hashtable()
    
    if D1.length != D2.length
        return "Not equal set of sequence"
    
    #assign default values as true 
    for i= 0 to D1.length
        assign D1[i] to true
    for i= 0 to D2.length
        assign D2[i] to true
     
    #check if all elements of D1 in D2
    for i = 0 to D1.length
        if D1[i] not in D2
            return "sequences are not same"
            
    #check if all elements of D2 in D1
    for i = 0 to D2.length
        if D2[i] not in D1
            return "sequences are not same"
    
    return "Sequences are same"
    
~~~

#### Runtime:

We can observe that every loop runs with a time complexity of O(n). 

Since there are 4 for loops total runtime is 4*O(n)
> O(4n) is equal to O(n)

Given n d-digit numbers in which each digit can take on up to k possible values, Radix Sort correctly sorts
these numbers in Θ(d(n + k)) time if the stable sort it uses takes Θ(n + k) time.

Here d takes the value of n<sup>2</sup> -1.

### Algorithm of Radix Sort:
~~~
RADIX_SORT(A,d)
    for i = 1 -> d
        use a stable sort to sort the array A on digit i
        *We will use count sort since it is a stable sort*
~~~

### Complete Algorithm:

~~~
COUNTING_SORT(D,B,k)
    INPUT: List of unsorted elements D,
           B is output array defined to store output list,
           k is max element in D
    OUTPUT:List of sorted elements in B
    /* len(D) = len(B), max(D) = k*/.
    Create a new array C[0..K]
    input_length = len(D)
    for i = 0 → k do
    C[i] = 0
    end
    for j = 1 → input_length do
    C[D[j]] = C[D[j]] + 1
    end
    for i = 1 → k do
    C[i] = C[i] + C[i − 1]
    end
    for j = input_length downto 1 do
    B[C[D[j]]] = D[j]
    C[D[j]] = C[D[j]] − 1
    end
~~~

~~~

# Problem 4: Merge-sort

Implement a bottom-up merge-sort for a collection of items by placing each item in its own
queue, and then repeatedly merging pairs of queues until all items are sorted within a single
queue.

```python
from collections import deque

##Define merge method, I took the same merge method from the Top-Down algo example provided. 
def merge(left_list, right_list):
    low = high = 0
    aux_list = []
    while low < len(left_list) and high < len(right_list):
        if left_list[low] < right_list[high]:
            aux_list.append(left_list[low])
            low += 1
        else:
            aux_list.append(right_list[high])
            high += 1

    # append what is remained in either of the lists
    aux_list.extend(left_list[low:])
    aux_list.extend(right_list[high:])
#     print(aux_list)
    return aux_list
    
def mergesort(ListA):
    ##First, break list into a list of lists e.g [[1],[4], ...]    
    
    ##Second, store lists in queue.
    queue = deque()
    for i in ListA:
        queue.append([i])

    ##Merge Sort Bottom-up method starts here.
    while(len(queue) > 1):  
    #Pop left and right lists from queue
        left_list = queue.popleft()
        right_list = queue.popleft()
    #Merge lists until there is one list left in the queue
        queue.append(merge(left_list,right_list))
    
    #Pop last remaining sorted list
    sorted_list = queue.pop()
    for i in range(len(ListA)):
        ListA[i]= sorted_list[i]
```

# Problem-5 : Heap Sort

Implement the heap-sort algorithm given in algorithm 1. The max_heapify and build_max_heap
procedures are described in algorithm 2 and algorithm 3, respectively.

```python
import matplotlib.pyplot as plt
import random as rn
import time


# initialise parent node
def parent(i):
    return i // 2

#Initialise left child
# Since using 0 based index returns 2(i+1)-1
def left_child(i):
    return (2 * (i + 1)) - 1

#Initialise right child
# Since using 0 based index returns2(i+1)
def right_child(i):
    return 2 * (i + 1)

#Heapify the list input at index i and given heapsize
def max_heapify(ListA, i, heap_size):
    left = left_child(i)
    right = right_child(i)
    if left <= heap_size and ListA[left] > ListA[i]:
        largest = left
    else:
        largest = i
    if right <= heap_size and ListA[right] > ListA[largest]:
        largest = right
    if largest != i:
        # print("before swap : ", ListA)
        ListA[i], ListA[largest] = ListA[largest], ListA[i]
        # print("swap happened? : ", ListA)
        max_heapify(ListA, largest, heap_size)

#build max heap 
def build_max_heap(ListA):
    A_heapsize = len(ListA) - 1
    for i in range(A_heapsize // 2, -1, -1):
        max_heapify(ListA, i, A_heapsize)

#call heapsort by combining above methods
def heapsort(ListA):
    build_max_heap(ListA)
    # print("after building max heap: ", ListA)
    n = len(ListA) - 1
    for i in range(len(ListA) - 1, 0, -1):
        ListA[0], ListA[i] = ListA[i], ListA[0]
        n = n - 1
        max_heapify(ListA, 0, n)
    return ListA
```

### Implementation:

Heap sort is a inplace algorithm which uses heap data structure.In this heapsort we use max heap to sort the elements. Heaps has nodes as Parent(i),Left_child(i),Right_child(i).Heapsort has two attributes: 
* Length of input
* Heap size

**_max_heapify()_:**
We use the max_heapify function which runs at node i to convert the node in correspondence to max heap. If the element at the node i is smaller than its children then we swap it with the higher value element with the parent.

**_Build_max_heap()_:**
The input needs to be built into max heap datastructure initially.It runs for half of length because the remaining nodes will be leaf nodes.

**_Heapsort()_:**
We need to first build max heap using _build-max-heap()_ Then we iterate over the input and Remove the root node and exchange it with the last node element and call max_heapify again to contruct the max heap structure.

Note: The implementation is 0 index based.

# Problem 6: Counting Sort
Implement the counting-sort algorithm given in algorithm 4.

```python
def counting_sort(inputA, outputB, k):
    input_length = len(inputA)
    Freq_list = [0 for i in range(0, k+1)]
    #to get the frequency of numbers in inputA
    for j in range(0, input_length):
        Freq_list[inputA[j]] = Freq_list[inputA[j]] + 1
    # update the index positions storing the various elements in outputB
    for i in range(1, k+1):
        Freq_list[i] = Freq_list[i] + Freq_list[i-1]
    # to store the elemnts in sorted way in outputB
    for m in range(input_length-1, -1, -1):
        outputB[Freq_list[inputA[m]]-1] = inputA[m]
        Freq_list[inputA[m]] = Freq_list[inputA[m]] - 1
    return outputB
```

Counting sort assumes that each of the n input elements is an integer 0 to k; for each k belongs to integer. Counting sort determines for each input element x the number of elements less than x.It uses this information to place element x directly into its position.


The _counting_sort()_ takes input array, output array which has length of input array and max element in the input array as input.
```python
import math

def Bucket_sort(input_list):
    
    input_length = len(input_list)

    #Create a list of buckets
    B = []
    for i in range(len(input_list)):
        B.append([])

    #Store the elements in buckets
    for i in range(0,len(input_list)):
        B[int(input_length*input_list[i])].append(input_list[i])

    #Insertion sort function to sort each bucket
    def insertion_sort(A):
        for i in range(0,len(A)):
            j = i - 1
            temp = A[i]
            while j >= 0 and temp < A[j]:
                A[j+1] = A[j]
                j -= 1
            A[j+1] = temp
        return A

    #Sort each list and append the lists
    sorted_list = []
    for i in range(len(input_list)):
        sorted_list = sorted_list + insertion_sort(B[i])
    return sorted_list
```

### Implementation:

Bucket sort works on the assumption that the input drawn is from the uniform distribution that is the input is generated by a random process that that distributes elements uniformly and independantly over a interval [0,1). 

The bucket sort algorithm divides the interval [0,1) into n equal sized buckets. Since the input data is uniformly distributed a single bucket wont have many numbers.After putting into buckets we sort the elements in each bucket usng insertion sort algorithm.Then we concatenate the data from various buckets sequentially which gives us a sorted output list.

### Runtime:

The average case runtime of Bucket sort is O(n).Worst case of bucket sort can be O(n<sup>2</sup>) if all the keys fall in the same bucket and happen to be in reverse order. If the data is not unifromly distributed then the growth of fucntion follows the runtime of Insertion sort.
